```python
rxns_w_probs = []
rxns_no_prob = []
rxns_not_in_probs_dic = []

for rxn in universal.reactions:
    try:
        if rxn_probs[rxn.id] > 0:
            rxns_w_probs.append(rxn.id)
        elif rxn_probs[rxn.id] == 0:
            rxns_no_prob.append(rxn.id)
    except:
        rxns_not_in_probs_dic.append(rxn.id)
```
